```python
city_realty_cleaned = city_realty.copy()
```

```python
full_address = city_realty_cleaned.Address.str.split(',', expand = True)
```

```python
full_address.columns = ['Address', 'Apartment Number']
```

```python
city_realty_cleaned = pd.concat([full_address, city_realty_cleaned[['Neighborhood','Beds', 'Size','Price/Ft2', 'Price', 'Date']]], axis = 1)
```

```python
city_realty_cleaned
```

```python
city_realty_cleaned[['Neighborhood', 'Second Neighborhood']] = city_realty['Neighborhood'].str.split('/', expand = True)
```

```python
city_realty_cleaned.loc[0].Address + " " + city_realty_cleaned.loc[0].Neighborhood + " NY"
```

```python
city_realty_cleaned['Address'] = city_realty_cleaned.Address.replace({'First':'1st', 'Second':'2nd', 'Third':'3rd', 'Fourth':'4th', 'Fifth':'5th', 
                                     'Sixth':'6th', 'Seventh':'7th', 'Eighth':'8th', 'Ninth':'9th'}, regex = True)
```

```python
city_realty_cleaned
```

```python
city_realty_cleaned['Address for Search'] = city_realty_cleaned['Address'] + ' ' + city_realty_cleaned['Neighborhood'] + ' NY'
```

```python
city_realty_cleaned
```

```python
geocode = RateLimiter(locator.geocode, min_delay_seconds=1)
```

```python
city_realty_cleaned['searched_loc'] = city_realty_cleaned['Address for Search'].apply(geocode)
```

```python
city_realty_cleaned['searched_loc'][0][-2].split(',')[-2].strip()
```

```python
city_realty_cleaned[city_realty_cleaned['searched_loc'].notnull()]
```

```python
location4 = geocode("39 East 29th Street New York County NY")
```

```python
location4
```

```python
city_realty_cleaned['Zip Code'] = city_realty_cleaned[city_realty_cleaned['searched_loc'].notnull()]['searched_loc'].apply(lambda x: x[-2].split(',')[-2].strip())
```

```python
city_realty_cleaned.Date = city_realty_cleaned.Date + '-2019'
```

```python
city_realty_cleaned
```

```python
city_realty_cleaned = city_realty_cleaned[['Address', 'Apartment Number', 'Neighborhood', 'Zip Code', 'Beds', 'Size', 'Price', 'Date']]
```

```python
city_realty_cleaned
```

```python
price_dict = {'K': 1000, 'M': 1000000}
```

```python
city_realty_cleaned['Price'] = [float(s[1:-1]) * price_dict[s[-1]] for s in city_realty_cleaned['Price']]
```
