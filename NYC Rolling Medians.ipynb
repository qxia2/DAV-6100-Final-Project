```python
import pandas as pd
```

```python
nyc_rolling = pd.read_csv('/Users/natanbienstock/Documents/Grad School/Fall 2019/DAV-6100/Final Project/nyc_rolling.csv')
```

```python
nyc_rolling.head()
```

```python
a = nyc_rolling['SALE DATE'].str.split('-', expand = True)
```

```python
a.columns = ['Year', 'Month', 'Day']
```

```python
a['Year-Month']= a['Year'] + '-' + a['Month']
```

```python
b = pd.merge(nyc_rolling,a, left_index = True, right_index = True)
```

```python
greater_than_zero = b[b['SALE PRICE'] > 0]
```

```python
d = greater_than_zero.groupby(['Year-Month', 'ZIP CODE'])['SALE PRICE'].median().reset_index(name = 'Median Sales')
```

```python
d.sort_values('ZIP CODE')
```

```python
d.to_csv('/Users/natanbienstock/Documents/Grad School/Fall 2019/DAV-6100/Final Project/nyc_rolling_medians.csv',
        index = False)
```
