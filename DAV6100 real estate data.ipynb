```python
import pandas as pd
import requests
url = 'https://data.cityofnewyork.us/resource/8vgb-zm6e.json'
resp = requests.get(url)
resp
```

```python
# now convert the 'resp' object's JSON content into a list of 
# native Python objects
data = resp.json()
type(data)
```

```python
# check the content of the list
data
```

```python
# now convert some of the extracted JSON data into a data frame: in this
# example we are extracting the 'number', 'title', 'labels' and 'state'
# components of the dict object
issues = pd.DataFrame(data, columns=['boro', 'block','lot','address_2','address_3','city_state_zip','original_market_value',
                                     'original_assessed_value', 'revised_market_value','revised_assessed_value','bbl'])
issues
```
