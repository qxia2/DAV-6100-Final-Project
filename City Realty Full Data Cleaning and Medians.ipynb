```python
search = SearchEngine()
```

```python
search.by_coordinates(40.698844,-73.99277 , radius = 10, returns = 1)[0].to_dict()['zipcode']
```

```python
c_s['lat,long'] = c_s['lat'].astype(str) + ',' + c_s['lng'].astype(str)
```

```python
c_s.head()
```

```python
def search_lat(lat_long):
    lat = float(lat_long.split(',')[0])
    long = float(lat_long.split(',')[1])
    return search.by_coordinates(lat, long, radius = 10, returns = 1)[0].to_dict()['zipcode']
```

```python
c_s['zip'] = c_s['lat,long'].apply(search_lat)
```

```python
city_realty_cleaned = c_s[['building_address', 'unit', 'hood', 'building_type', 'bedrooms', 'bathrooms','sale_price', 'sale_date','zip']]
```

```python
city_realty_cleaned.head()
```

```python
city_realty_cleaned.to_csv('/Users/natanbienstock/Documents/Grad School/Fall 2019/DAV-6100/Final Project/city_realty_full_cleaned.csv'
                     , index = False)
```

```python
city_realty_dates = city_realty_cleaned['sale_date'].str.split('/', expand=True)
```

```python
city_realty_dates.columns = ['Year', 'Month', 'Day']
```

```python
city_realty_dates['Year-Month'] = city_realty_dates['Year'] + '-' +city_realty_dates['Month']
```

```python
city_realty_dates.head()
```

```python
city_realty_medians = pd.merge(city_realty_cleaned, city_realty_dates, left_index = True, right_index = True)
```

```python
city_realty_medians.head()
```

```python
city_realty_medians = city_realty_medians.groupby(['Year-Month', 'zip'])['sale_price'].median().reset_index(name = 'Median Sales')
```
